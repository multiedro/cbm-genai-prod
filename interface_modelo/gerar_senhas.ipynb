```python
import bcrypt

# Lista de senhas (todas iguais neste exemplo)
senhas = ['cbm12345'] * 10

# Lista de e-mails
emails = [
    'danilo@cbmlegal.com.br',
    'flavio@cbmlegal.com.br',
    'filipe@cbmlegal.com.br',
    'stefano@cbmlegal.com.br',
    'mariafernanda@cbmlegal.com.br',
    'beatriz@cbmlegal.com.br',
    'manoela@cbmlegal.com.br',
    'nathalia@cbmlegal.com.br',
    'administrativo@cbmlegal.com.br',
    'maria.victoria@cbmlegal.com.br'
]

# Gerar hashes manualmente com bcrypt
for email, senha in zip(emails, senhas):
    usuario = email.split('@')[0].replace('.', '')
    nome = usuario.capitalize()
    hash_senha = bcrypt.hashpw(senha.encode(), bcrypt.gensalt()).decode()

    print(f"""    {usuario}:
      email: {email}
      first_name: {nome}
      last_name: null
      logged_in: false
      password: {hash_senha}
      password_hint: Gerado pela Multiedro
      roles: null""")

```
